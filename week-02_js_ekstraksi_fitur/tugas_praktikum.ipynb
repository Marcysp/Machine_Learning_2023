### 1. Seleksi variabel. Memisahkan antara variabel yang dapat digunakan dan tidak dapat digunakan.

```python
import pandas as pd 

# variabel 'data' merupakan path dari file csv yang dituju
data = 'data/wbc.csv'
# Baca data dari file CSV
df = pd.read_csv(data)

# kita memiliki file CSV dalam variabel df
# Pemisahan variabel

# variabel yang tidak dapat digunakan
unusable_col = ['id']

#variabel yang dapat digunakan
usable_col = [col for col in df.columns if col not in unusable_col]

# pisahkan data berdasarkan variabel yang dapat digunakan dan tidak
data_unuse = df[unusable_col]
data_use = df[usable_col]

print("Variabel yang tidak dapat digunakan :")
print(data_unuse.head())

print("Variabel yang dapat digunakan :")
print(data_use.head())
```

penjelasa : Menampilkan variabel dataset wbc yang variabelnya dapagt digunakan dan tidak dapat digunakan

### 2. Lakukan proses encoding pada kolom "diagnosis"


```python
#import library pemodelan statistik seperti encoding
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()# membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding

#cek hasil dengan menampilkan kolom diagnosis hasil encoding
print(df['diagnosis'].head())
df.head()
```

penjelasan : menampilkan data kolom diagnosis setelah di encoding. dapat dilihat pada kolom diagnosis yang awalnya MM berubah menjadi angka 1

### 3. Melakukan proses standarisasi pada semua kolom yang memiliki nilai numerik

```python
std = StandardScaler()# membuat objek dari StandardScaler

#memilih data numerik dari data yang dapat digunakan (data_use)
numeric_cols = data_use.select_dtypes(include=['float64', 'int64']).columns

#standarisasi kolom numerik
df[numeric_cols] = std.fit_transform(df[numeric_cols])

# tampilkan beberapa baris pertama setelah standarisasi untuk memastikan
df[usable_col].head()
```

penjelasan : Melakukan standarisasi pada kolom-kolom dataset wbc yang variabelnya dapat digunakan dan memiliki nilai numerik. jadi pada kolom diagnosis hanya tampil hasil setelah di encoding namun tidak dilakukan standarisasi pada kolom diagnosis karena buka merupakan kolom numerik.

### 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20

```python
from sklearn.model_selection import train_test_split

# Split data latih dan data uji denagn rasio 80:20
data_latih, data_uji = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# cek ukuran data
print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data latih:\n{data_latih.diagnosis.value_counts()}')
print(f'Jumlah label data uji:\n{data_uji.diagnosis.value_counts()}')
```

penjelasan : tampil hasil setelah di split dengan rasio yang telah di tentukan sehingga jumlah label data latih lebih banyak dibandingkan dengan label data uji seperti yang di minta.
